```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.chdir("D:/Pandas")

car=pd.read_csv("Toyota.csv", index_col=0,na_values=["??","????"])	#set first column as index, remove ??, ???? values
car.dropna(axis=0,inplace=True)

```

```python
#Scatter Plot Age vs Price
sns.regplot(x = car['Age'], y = car['Price'],color = 'orange')
```

```python
#Dark Background
sns.set(style = 'darkgrid')
sns.regplot(x = car['Age'], y = car['Price'],color = 'green')
```

```python
#Without regression Fit line
sns.regplot(x = car['Age'], y = car['Price'], fit_reg = False )
```

```python
#USE Hue
sns.lmplot(x = 'Age',y='Price',data = car, fit_reg = False, hue = 'FuelType',
           legend = 'True', palette = 'Set1')
```

```python
#Histogram

sns.displot(car['Age'])
```

```python
#BarGraph
sns.countplot(x= 'FuelType', data = car)
```

```python
#With Hue
sns.countplot(x= 'FuelType', data = car, hue = 'Automatic')
```

```python
#Line
sns.lineplot(x= 'FuelType',y = 'Price', data = car)  # Plot the chart
```

```python
#boxplot
sns.boxplot(y = 'Age', data = car)
```

```python
#heatmap
hm = sns.heatmap(car.corr())
```

```python
sns.pairplot(car, vars = ['Age','Price','KM'])
plt.show()
```

```python
graph = sns.FacetGrid(car, col = 'FuelType', row = 'MetColor')
# map the above form facetgrid with some attributes
graph.map(plt.scatter,"Price", "Age")
# show the object
plt.show()
  
```
