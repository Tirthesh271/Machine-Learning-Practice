```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
os.chdir("D:/Pandas")

car=pd.read_csv("Toyota.csv", index_col=0,na_values=['??','????'])	#set first column as index, remove ??, ???? values
car.dropna(axis=0,inplace=True)
```

```python
#Scatter plot using pandas
car.plot(kind = 'scatter',
        x = 'Age',
        y = 'Price',
        color = 'Yellow')
  
# set the title
plt.title('ScatterPlot')
  
# show the plot
plt.show()
```

```python
# bar plot
car[:20]['Age'].plot.bar()
```

```python
#Histogram
car['KM'].plot.hist()
```

```python
#Line Chart
car.plot.line(x="KM",y = "Price")
```

```python
#Box chart
car['CC'].plot.box()
```

```python
#Heatmap
corr = car.corr()
corr.style.background_gradient()
```
