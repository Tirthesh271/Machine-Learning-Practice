```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
os.chdir("D:/Pandas")

car=pd.read_csv("Toyota.csv", index_col=0,na_values=['??','????'])	#set first column as index, remove ??, ???? values
car.dropna(axis=0,inplace=True)
```

```python
plt.scatter(car['Age'], car['Price'], c='red' )
plt.title("My Plot")
plt.xlabel("Age in Months")
plt.ylabel("Price in Euros")
plt.show()
```

```python
plt.hist(car['KM'],color='red',edgecolor = 'white',bins=5)
plt.title("My Plot")
plt.xlabel("KM")
plt.ylabel("Frequency")
plt.show()
```

```python
car['FuelType'].value_counts()
counts = [968,116,12]
FT = ('Petrol','Diesel','CNG')
index = np.arange(len(FT))

plt.bar(index,counts,color = ['red','blue','green'])
plt.title("My bar graph")
plt.xlabel("Fuel Type")
plt.ylabel("Freq")
plt.xticks(index,FT,rotation = 90)
plt.show()
```

```python
#Line chart

x = np.array([1, 2, 3, 4])  # X-axis points
y = x*2  # Y-axis points
  
plt.plot(x, y)  # Plot the chart
plt.show()  # display
'''

plt.plot(car['Age'],car['Price'])
plt.title("My Line graph")
plt.xlabel("")
plt.ylabel("")
plt.show()

'''
```
