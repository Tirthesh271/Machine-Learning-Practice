```python
import numpy as np
import pandas as pd
import os

os.chdir("D:/pandas")

car = pd.read_csv("Toyota.csv",index_col = 0)
car
```

```python
try:car.insert(10,'Price_Class',"")
except:pass
for i in range(0,len(car['Price']),1):
    if (car["Price"][i]<=8450):
        car['Price_Class'][i]='Low'
    elif (car["Price"][i]>11950):
        car['Price_Class'][i]='High'
    else:
        car['Price_Class'][i] = 'Medium'
car['Price_Class'].value_counts()

```

```python
try:car.insert(11,'Age_Converted',0)
except:pass

def m2y(val):
    years = val/12
    return years
car['Age_Converted'] = m2y(car['Age'])
print(car[['Age','Age_Converted']])
```

```python
try:car.insert(12,'KM_Per_Month',0)
except:pass

def KmRun(val1,val2):
    ratio = val2/val1
    return ratio

car['KM'].replace("??",1,inplace=True)#remove ??
car['Km_Per_Month'] = KmRun(car['Age'],car['KM'].astype(int))#covert str to int
print(car[['Age','Km_Per_Month']])
```

```python
car.to_csv("D:/Pandas/Toyota_Exported.csv")
```

```python
try:car.insert(13,'Worth_To_Buy','')
except:pass

for i in range(0,len(car['Price_Class']),1):
    if (car['Price_Class'][i]== 'Medium') and (car['KM_Per_Month'][i]<2000) and (car['Price'][i]<12000):
        car['Worth_To_Buy'][i] = ' Yes'
    else:car['Worth_To_Buy'][i] = ' No'
car
        
```
